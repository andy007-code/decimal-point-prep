## 2. Basic Arithmetic
Python supports all standard mathematical operations. Understanding operator precedence (BODMAS) is important — Python follows the same rules as mathematics.

| Operator | Meaning | Example |
|---|---|---|
| `+` | Addition | `5 + 3 = 8` |
| `-` | Subtraction | `5 - 3 = 2` |
| `*` | Multiplication | `5 * 3 = 15` |
| `/` | Division | `5 / 3 = 1.667` |
| `**` | Power / Exponent | `5 ** 3 = 125` |
| `//` | Floor Division | `5 // 3 = 1` |
| `%` | Modulo (remainder) | `5 % 3 = 2` |

```python
print(5 + 3)
print(5 - 3)
print(5 * 3)
print(5 / 3)
print(5 ** 3)   # 5 to the power of 3
print(5 // 3)   # integer division — drops the decimal
print(5 % 3)    # remainder after division
```

### Financial Application
Arithmetic operations are the foundation of every financial formula. For example, calculating the value of an investment after one year of growth:

```python
principal = 100000    # Rs 1 lakh invested
rate = 0.085          # 8.5% annual interest rate
years = 1

final_value = principal * (1 + rate) ** years
print(f'Investment grows to: Rs {final_value:,.2f}')
```
