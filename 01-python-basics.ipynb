## 1. Variables and Data Types
A **variable** is a named container that stores a value. Python automatically detects the type of data you store — you don't need to declare types explicitly.

The four most common data types:
- `int` — whole numbers (e.g. number of shares)
- `float` — decimal numbers (e.g. stock price, returns)
- `str` — text (e.g. company name, ticker symbol)
- `bool` — True or False (e.g. is the market open?)

````python
# Integer — whole number
shares = 100

# Float — decimal number
price = 2450.75

# String — text
ticker = 'RELIANCE.NS'

# Boolean — True or False
market_open = True

print(type(shares))       # <class 'int'>
print(type(price))        # <class 'float'>
print(type(ticker))       # <class 'str'>
print(type(market_open))  # <class 'bool'>
````

````python
# Variables can be used in calculations
portfolio_value = shares * price
print(f'Portfolio Value: Rs {portfolio_value:,.2f}')
````

## 2. Functions
A **function** is a reusable block of code that performs a specific task. Instead of rewriting the same logic multiple times, you define it once and call it whenever needed.

Syntax:
```python
def function_name(parameter1, parameter2):
    # code here
    return result
```
- `def` keyword defines the function
- Parameters are inputs the function receives
- `return` sends a value back to the caller
- Indentation (4 spaces) defines what's inside the function

````python
def simple_interest(principal, rate, time):
    """
    Calculate simple interest.
    principal : initial amount invested (Rs)
    rate      : annual interest rate as a decimal (e.g. 0.08 for 8%)
    time      : number of years
    Returns   : interest earned (Rs)
    """
    interest = principal * rate * time
    return interest

# Test the function
result = simple_interest(100000, 0.08, 3)
print(f'Simple Interest earned: Rs {result:,.2f}')
````

## 3. Conditionals (if / elif / else)
Conditionals allow the program to make decisions. Code inside an `if` block runs only when the condition is True.

**Comparison operators:**
- `>` greater than, `<` less than
- `>=` greater than or equal, `<=` less than or equal
- `==` equal to, `!=` not equal to

**Logical operators:**
- `and` — both conditions must be True
- `or` — at least one condition must be True
- `not` — reverses True/False

````python
daily_return = 2.5   # percentage

if daily_return > 2:
    print('Big Gain — market rallied strongly')
elif daily_return > 0:
    print('Small Gain — market edged higher')
elif daily_return == 0:
    print('Flat — no change')
elif daily_return > -2:
    print('Small Loss — market dipped slightly')
else:
    print('Big Loss — market sold off heavily')
````

## 4. Lists
A **list** is an ordered collection of items. Lists are defined using square brackets `[]` and can hold any data type — even a mix.

Key properties:
- **Ordered** — items stay in the order you put them
- **Mutable** — you can change, add, or remove items
- **Zero-indexed** — the first item is at index 0, not 1
- **Negative indexing** — index -1 gives the last item

````python
# A list of stock prices over 7 days
prices = [100, 102, 98, 105, 110, 108, 112]

print(prices[0])    # first price — index 0
print(prices[-1])   # last price — negative index
print(prices[2:5])  # slice — items at index 2, 3, 4 (stops before 5)
````

````python
# Common list methods
prices.append(115)    # add item to end
print(prices)

prices.pop()          # remove last item
print(prices)

print(len(prices))    # number of items in the list
print(sorted(prices)) # returns a sorted copy (does not change original)
````

## 5. Loops
A **loop** repeats a block of code multiple times, avoiding repetitive code.

- `for` loop — iterates over each item in a sequence (list, range, string)
- `while` loop — repeats as long as a condition is True
- `break` — exits the loop immediately
- `continue` — skips the current iteration and moves to the next

`range(start, stop, step)` generates a sequence of numbers. `range(1, 6)` gives 1, 2, 3, 4, 5.

````python
# For loop — compute cumulative return
prices = [100, 102, 98, 105, 110]

for i in range(len(prices)):
    print(f'Day {i+1}: Price = {prices[i]}')
````

````python
# While loop — find first day price crosses 108
day = 0
while prices[day] < 108:
    day += 1
print(f'Price crossed 108 on Day {day + 1} at Rs {prices[day]}')
````

## 6. Strings
A **string** is a sequence of characters enclosed in quotes. In financial data work, strings are used for ticker symbols, company names, dates, and cleaning messy datasets.

Strings are **immutable** — you cannot change a character in place; you create a new string instead.

````python
ticker = '  reliance.ns  '

print(ticker.strip())            # remove leading/trailing spaces
print(ticker.strip().upper())    # clean and convert to uppercase
print(ticker.strip().replace('.ns', ''))  # remove exchange suffix
````

````python
# Splitting a date string — useful when dates come as text in CSV files
date = '2024-01-15'
parts = date.split('-')
print(f'Year: {parts[0]}, Month: {parts[1]}, Day: {parts[2]}')
````
