Pandas is the primary data analysis library in Python. Built on top of NumPy,
it provides two core data structures — Series and DataFrame — that make working
with tabular financial data fast and intuitive. Every financial dataset (stock
prices, returns, macro indicators) will be handled using Pandas.

```python
import pandas as pd
import numpy as np
```

## 1. Creating and Reading Data

A **DataFrame** is a 2-dimensional table with labelled rows and columns —
think of it as an Excel sheet inside Python. A **Series** is a single column
from that table.

Two ways to create a DataFrame:
- Manually from a dictionary
- Reading from a CSV file using pd.read_csv()

```python
# Creating a DataFrame manually from a dictionary
data = {
    'Ticker':  ['RELIANCE', 'TCS', 'INFY', 'HDFCBANK', 'ITC'],
    'Price':   [2450, 3800, 1500, 1600, 420],
    'Sector':  ['Energy', 'IT', 'IT', 'Banking', 'FMCG'],
    'PE_Ratio':[22.5, 28.3, 24.1, 18.7, 26.4]
}

df = pd.DataFrame(data)
print(df)
```

```python
# A Series is just one column of a DataFrame
prices = df['Price']
print(type(prices))   # confirms it's a Series
print(prices)
```

```python
# Key DataFrame inspection commands
# Always run these first when you load any new dataset

print(df.shape)        # (rows, columns)
print(df.dtypes)       # data type of each column
print(df.head(3))      # first 3 rows
print(df.tail(2))      # last 2 rows
print(df.info())       # summary — nulls, types, memory
```

## 2. Indexing, Selecting and Assigning

Two ways to select data from a DataFrame:
- **iloc** — select by position (integer index, like a list)
- **loc** — select by label (row/column name)

Key difference in slicing:
- iloc[0:3] gives rows 0, 1, 2 — stops before 3
- loc[0:3] gives rows 0, 1, 2, 3 — includes 3

```python
# iloc — selecting by position
print(df.iloc[0])         # first row
print(df.iloc[0, 1])      # first row, second column (Price)
print(df.iloc[0:3])       # first 3 rows
```

```python
# loc — selecting by label
print(df.loc[0, 'Ticker'])         # row 0, Ticker column
print(df.loc[0:2, 'Price'])        # rows 0 to 2 (inclusive), Price column
print(df.loc[df['Sector'] == 'IT']) # all IT stocks — boolean filtering
```

```python
# Adding a new column — Market Cap estimate
df['MarketCap_Cr'] = df['Price'] * 100  # simplified estimate
print(df)
```

## 3. Summary Functions and Maps

Summary functions condense a DataFrame into key statistics.
df.describe() is the first command to run on any new financial dataset —
it instantly shows the range, average, and spread of every numeric column.

map() and apply() transform values across a Series or DataFrame.
- map()   — transforms each value in a Series using a function or dictionary
- apply() — applies a function along rows or columns of a DataFrame

```python
# describe() — instant statistical summary of all numeric columns
print(df.describe())
```

```python
# Individual statistics on a column
print(f"Mean Price:   {df['Price'].mean():,.2f}")
print(f"Max PE Ratio: {df['PE_Ratio'].max()}")
print(f"Min Price:    {df['Price'].min():,.2f}")
print(f"Std Dev:      {df['Price'].std():,.2f}")

# value_counts — how many stocks per sector
print("\nStocks per Sector:")
print(df['Sector'].value_counts())
```

```python
# map() — label each stock based on PE ratio
# PE > 25 = expensive, PE < 25 = fair value (simplified rule)
df['Valuation'] = df['PE_Ratio'].map(
    lambda x: 'Expensive' if x > 25 else 'Fair Value'
)
print(df[['Ticker', 'PE_Ratio', 'Valuation']])
```

## 4. Grouping and Sorting

groupby() splits the DataFrame into groups based on a column value
and applies a function to each group. It answers questions like:
"What is the average PE ratio per sector?" or "Which sector has the
highest average price?"

sort_values() sorts rows by any column — ascending or descending.
agg() lets you apply multiple functions to multiple columns at once
after a groupby.

```python
# Average price and PE ratio per sector
sector_stats = df.groupby('Sector')[['Price', 'PE_Ratio']].mean()
print(sector_stats)
```

```python
# agg() — multiple functions at once
sector_agg = df.groupby('Sector')['Price'].agg(['mean', 'max', 'min'])
print(sector_agg)
```

```python
# sort_values — rank stocks by PE ratio highest to lowest
print(df[['Ticker', 'Sector', 'PE_Ratio']].sort_values('PE_Ratio', ascending=False))
```
